```python
import ipywidgets as widgets
```

```python
c=widgets.Dropdown(
    options=["rain","snow"],
    value='rain',
    description='Graph:',
    disabled=False,
)

def graph(a):
  sns.barplot(y="Month",x="temperature",data=df[df["Precip Type"]==a],estimator=len)
  plt.show()
```

```python
widgets.interact(graph,a=c)
```
