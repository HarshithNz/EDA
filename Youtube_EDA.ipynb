13. Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.

```python
sns.lineplot(x="view_count",y="dislikes",data=df,estimator="sum")
```

```python
sns.scatterplot(x="view_count",y="dislikes",data=df)
```

```python
sns.heatmap(df.corr(),annot=True,cmap='Blues') #dislikes and view_count has correlation of 0.68 its moderately related.
```

**Inferences**


*   **Lineplot :** lineplot shows patterns and trends which tells that view_count has correlation with dislikes rates.
*   **Scatterplot :** The scatterplot displays data points representing the number of dislikes ("dislikes") against the number of views ("view_count") for a set of videos.
*   **Heatmap Description :** The heatmap visualizes the relationship between the number of views ("view_count") and the number of dislikes ("dislikes") for a dataset of videos. A correlation coefficient of 0.68 between "view_count" and "dislikes" indicates a moderately strong positive correlation. This positive correlation suggests that as the number of views increases, there is a tendency for the number of dislikes to also increase.
*   The correlation between views and dislikes is [0.68] which can be considered as moderate to higher correlation. So, from this we can conclude that views have a significance in determining the dislikes of a video.
